```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

X = np.array([11,5,10,12,7], dtype=float)
y = np.array([7,4,5,8,6], dtype=float)

modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mse'
)
```

```python
historial = modelo.fit(X, y, epochs=2000, verbose=False)
```

```python
print(modelo.layers[0].get_weights())
```

```python
print(modelo.predict(np.array([20])))
```

```python
plt.scatter(X, y, label='Datos originales')
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dispersión de datos y línea de regresión')
plt.legend()
plt.show()
```
