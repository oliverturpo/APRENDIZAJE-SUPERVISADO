```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor, plot_tree

# Datos
X = np.array([70, 85, 55, 62, 90, 88, 100, 98]).reshape(-1, 1)
y = np.array([22, 30, 28, 35, 40, 55, 64, 60])


```

```python
arbol = DecisionTreeRegressor(max_depth=8, random_state=30)
arbol.fit(X,y)
```

```python
plt.figure(figsize=(12,8))
plot_tree(arbol, feature_names=['Ingresos'], filled=True)
plt.show()
```

```python
x_nuevo = np.array([[99]])
y_pred = arbol.predict(x_nuevo)
print(y_pred)
```
